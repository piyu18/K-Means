```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
plt.figure(figsize=(8,5))
X1 = [3, 1, 1, 2, 1, 7, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8]
X2 = [3, 1, 2, 2, 1, 6, 5, 6, 5, 6, 7, 8, 9, 8, 6, 7, 8]
plt.scatter(X1,X2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
d={'X1':X1, 'X2':X2}
df = pd.DataFrame(data=d)
plt.figure(figsize=(8,5))
clusters = range(2, 10)
silhouette_scores = []
for cluster in clusters:
 
 # initialise kmeans
 kmeans = KMeans(n_clusters=cluster)
 kmeans.fit(df)
 cluster_labels = kmeans.labels_
 
 # silhouette score
 silhouette_scores.append(silhouette_score(df, cluster_labels))

plt.plot(clusters,silhouette_scores,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette Analysis for Optimal Value of k')
plt.show() 
```
